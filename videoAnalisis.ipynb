```python
videos_category_trend = df["videoCategoryLabel"].value_counts()
videos_category_trend[0]
```

```python
plt.figure(figsize=(40, 15))
plt.title("Videos por categora", fontsize=30)
plt.rcParams.update({'xtick.labelsize': 10})
plt.xlabel("Categora", fontsize=20)
plt.bar(df["videoCategoryLabel"].unique(), videos_category_trend)
```

```python
views = df[[ "videoCategoryLabel", "viewCount"]].groupby([ "videoCategoryLabel"]).sum().sort_values(by="viewCount", ascending=False)
```

```python
views.assign(percentage=round(views["viewCount"]/views["viewCount"].sum()*100, 2))
```

```python
color = sns.color_palette("Set2", 10)
plt.figure(figsize=(10, 5))
plt.pie(views["viewCount"], labels=views.index, colors=color, autopct='%1.1f%%', shadow=True, startangle=90)
plt.show()
```

```python
df["publishedAt"] = pd.to_datetime(df["publishedAt"])
```

```python
dat = df["publishedAt"].dt.year.value_counts().sort_index()
dat.index
```

```python
plt.figure(figsize=(15, 5))
plt.rcParams.update({'xtick.labelsize': 20})
plt.bar(dat.index, dat)
plt.xticks(dat.index)
plt.show()
```

```python
# Categorias mas vistas por ao
year = df.groupby(df["publishedAt"].dt.year)[["videoCategoryLabel", "viewCount"]].value_counts()
year = year.to_frame()
year.reset_index(inplace=True)
year = year.groupby(["publishedAt", "videoCategoryLabel"]).sum().sort_values(by="publishedAt")
year.drop(columns=0, inplace=True)
year
```

```python
plt.figure(figsize=(15, 15))
sns.barplot(x="viewCount", y="videoCategoryLabel", hue="publishedAt", data=year.reset_index())

```
